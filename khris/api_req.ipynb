```python
import pandas as pd
import os
import csv
import requests


# Join Netflix_titles.csv
csvpath1 = os.path.join('..','Netflix Project', 'netflix_titles.csv')
with open(csvpath1, 'r', encoding='utf-8') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    for row in csvreader:
        print(row)


```

```python
netflix = pd.read_csv(csvpath1)
print("Null or NaN values")
print(netflix.isnull().sum())
```

```python
API_KEY = 'cc5b24bf'

def search_movie_country(movie_title):
    url = f'http://www.omdbapi.com/?apikey={API_KEY}&t={movie_title}'
    response = requests.get(url)
    data = response.json()
    if 'Country' in data:
        return data['Country']
    return None

# Filter country rows
missing_country_df = netflix[netflix['country'].isna()]

# Extract Titles [Movies / TV Shows] that have NaN
missing_country_movies = missing_country_df[missing_country_df['type'] == 'Movie']['title'].tolist()
missing_country_shows = missing_country_df[missing_country_df['type'] == 'TV Show']['title'].tolist()

# New empty DataFrame to store the new data
new_data = []

# API request to fetch country information for missing movies
for movie_title in missing_country_movies:
    country = search_movie_country(movie_title)
    if country:
        new_data.append({'title': movie_title, 'country': country})

# API request to fetch country information for missing TV shows
for show_title in missing_country_shows:
    country = search_movie_country(show_title)
    if country:
        new_data.append({'title': show_title, 'country': country})

# Convert the new data into a DataFrame
new_df = pd.DataFrame(new_data)
# Missing values in 
```

```python
# Confirm there are no NaN/Null in the newdf

print(new_df.isnull().sum())
```

```python
# Merge `updated_df` with `new_df` on 'title' column
updated_df = updated_df.merge(new_df, on='title', how='left', suffixes=('', '_new'))

# Drop the extra 'country_new' column
updated_df.drop('country_new', axis=1, inplace=True)

updated_df
```
